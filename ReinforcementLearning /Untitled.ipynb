```python
import numpy as np
import matplotlib.pyplot as plt 

data = np.loadtxt("mc_tile_returns.txt", dtype=float)

plt.plot(data,label="return")

meanv = 0
meanr = []
xr = []
for i, x in enumerate(data):
    meanv += x
    if i >= 24:
        meanr.append(meanv/25)
        meanv = meanv - data[i-24]
        xr.append(i+1)
# print(meanr)
plt.plot(xr,meanr,label="rolling mean")
plt.xlabel('Number of episodes')
plt.ylabel('return & mean return')
plt.legend()
plt.show()
```
